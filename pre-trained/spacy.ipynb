```python
import spacy
```

```python

nlp = spacy.load("ko_core_news_lg")
```

```python
text = """
김민수는 서울대학교에 재학 중이며
전화번호는 010-1234-5678이고
이메일은 minsu@gmail.com 입니다.
"""
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
