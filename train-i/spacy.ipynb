```python
#보호 하고 싶ㅇㄴ 데이터 지정
ENTITY_DEFINITIONS = [
    ("프로젝트 알파", "PROJECT_NAME"),
    ("2026년 6월 출시", "LAUNCH_TIME"),
    ("프로젝트 네오", "PROJECT_NAME"),
    ("SNS 마케팅 전략", "MARKETING_METHOD"),
    ("15억 원", "PROJECT_BUDGET"),
    ("프로젝트 오로라", "PROJECT_NAME"),
    ("TV 광고 중심의 마케팅 전략", "MARKETING_METHOD"),
]
```

```python
#preparing training data
#load data, label to training data 

training_data = []
entities = []

for ent_text, label in ENTITY_DEFINITIONS:
    start = text.find(ent_text)
    if start == -1:
        print(f"Entity not found: {ent_text}")
        continue

    end = start + len(ent_text)
    entities.append((start, end, label))

training_data.append({
    "text": text,
    "entities": entities
})

print(training_data)
```

```python
#train use docbin
import spacy
from spacy.tokens import DocBin
from spacy.util import filter_spans

nlp = spacy.blank("ko")   # PURE spaCy, no MeCab, label
nlp.tokenizer = spacy.tokenizer.Tokenizer(nlp.vocab)

doc_bin = DocBin()

for example in training_data:
    doc = nlp.make_doc(example["text"])
    ents = []

    for start, end, label in example["entities"]:
        span = doc.char_span(
            start,
            end,
            label=label,
            alignment_mode="contract"
        )
        if span:
            ents.append(span)

    doc.ents = filter_spans(ents)
    doc_bin.add(doc)

doc_bin.to_disk("train.spacy")

```
